# Analyzing Toronto Power Demand Data (2016-2018)



```python
%matplotlib inline

import sys
import statsmodels as ss
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Read the demand data for past 3 years
df =  pd.read_csv("data/power_data.csv", parse_dates=['Date'])
df.head()
```

```python
df1 = df[["Date", "Hour", "Toronto"]].where(df["Date"].dt.year == 2018)
df1.head()
df1 = np.isfinite(df1["Toronto"])

df1["Toronto"].plot()
plt.title("Toronto Demand")
plt.ylabel("Toronto")
plt.xlabel("Hour")
print(df1)
```

```python
df1.describe()
```

```python
ss_decomposition = seasonal_decompose(x=df1["Hour"], model='additive', freq=6)
estimated_trend = ss_decomposition.trend
estimated_seasonal = ss_decomposition.seasonal
estimated_residual = ss_decomposition.resid
```

```python
fig, axes = plt.subplots(4, 1, sharex=True, sharey=False)
fig.set_figheight(10)
fig.set_figwidth(15)

axes[0].plot(df1["Toronto"], label='Original')
axes[0].legend(loc='upper left');

axes[1].plot(estimated_trend, label='Trend')
axes[1].legend(loc='upper left');

axes[2].plot(estimated_seasonal, label='Seasonality')
axes[2].legend(loc='upper left');

axes[3].plot(estimated_residual, label='Residuals')
axes[3].legend(loc='upper left');
```

```python
pd.Series.rolling(df1["Toronto"], window=25).mean().plot(style="-g")
plt.title("Toronto Rolling Mean")
plt.ylabel("Toronto")
plt.xlabel("Hour")
```

```python
plt.figure()
df1["Toronto"].plot(kind="hist", stacked=True, bins=50)
plt.title("Toronto Power Demand histogram")
plt.xlabel("Power Demand")
```
