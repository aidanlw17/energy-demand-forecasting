```python
#power_df[(start<=power_df.index) & (power_df.index<=end)].plot(grid=True)
#print(power_df.head())
power = power_df.drop('Bruce',axis=1)
```

```python
#print(power.head())
print(power.shape[1])
sta=datetime(2016,1,1)
print(sta)
print(sta.year)
```

```python
#plotting by year
zones = list(power)
zones = zones[2:11]
for i in zones:
    t=0
    #print('Zone: ' + i)
    zone_col = power[[i]] 
    zone_col['Rolling Mean'] = zone_col[i].rolling(window=80).mean()
    zone_col['Max'] = zone_col[i].rolling(window=80).max()
    start=datetime(2016,1,1)
    end=datetime(2016,12,31)
    while t<3:
        #print("Interval is: " + str(start) + ' to: ' + str(end))
        zone_col[(start<=zone_col.index) & (zone_col.index<=end)].plot(grid=True)
        plt.ylabel('Power (GW)')
        plt.title('Power data by year in ' + str(start.year) + ' for ' + str(i))
        t+=1
        start=datetime(start.year+1,1,1)
        end=datetime(end.year+1,12,31)
    
    plt.close
        
```

```python
#plotting all zones power by month in the year
def plotAllZonesAllMonths(year):
    zones = list(power)
    zones = zones[2:11]
    for i in zones:
        t=0
        #print('Zone: ' + i)
        zone_col = power[[i]] 
        zone_col['Rolling Mean'] = zone_col[i].rolling(window=150).mean()
        zone_col['Max'] = zone_col[i].rolling(window=150).max()
        start=datetime(int(year),1,1)
        end=datetime(int(year),2,1)
        while end.month<12:
            #print("Interval is: " + str(start) + ' to: ' + str(end))
            zone_col[(start<=zone_col.index) & (zone_col.index<=end)].plot(grid=True)
            plt.ylabel('Power (GW)')
            plt.title('Power data in month ' + str(start.month) + ' in ' + str(start.year) + ' for ' + str(i))
            if(end.month==12):
                start=datetime(start.year,start.month+1,1)
                end=datetime(end.year,12,31)
            else:
                start=datetime(start.year,start.month+1,1)
                end=datetime(end.year,end.month+1,1)
        start=datetime(start.year+1,start.month,1)
        end=datetime(end.year+1,end.month,1)

        plt.close
        
#Monthly plots through the given year for given zone
def plotAllMonths(year,zone):
    i=str(zone)
    zone_col = power[[i]] 
    zone_col['Rolling Mean'] = zone_col[i].rolling(window=150).mean()
    zone_col['Max'] = zone_col[i].rolling(window=150).max()
    start=datetime(int(year),1,1)
    end=datetime(int(year),2,1)
    t=0
    while t<13:
        zone_col[(start<=zone_col.index) & (zone_col.index<=end)].plot(grid=True)
        plt.ylabel('Power (GW)')
        plt.title('Power data in month ' + str(start.month) + ' in ' + str(start.year) + ' for ' + str(i))
        if(end.month==12):
            if(start.month<12):
                start=datetime(start.year,start.month+1,1)
            end=datetime(end.year,12,31)
        if(start.month==12):
            start=datetime(start.year,start.month,1)
            end=datetime(end.year,end.month,31)
        else:
            start=datetime(start.year,start.month+1,1)
            end=datetime(end.year,end.month+1,1)
        t+=1

    plt.close
        
#Plotting one week of month from start day
def plotWeek(year, zone, month, startDay):
    i=str(zone)
    hours=power['Hours']
    zone_col = power[[i]] 
    zone_col['Rolling Mean'] = zone_col[i].rolling(window=5).mean()
    zone_col['Max'] = zone_col[i].rolling(window=5).max()
    start=datetime(int(year),int(month),int(startDay))
    end=datetime(int(year),int(month),int(startDay) + 6)
    t=0
    zone_col[(start<=zone_col.index) & (zone_col.index<=end)].plot(grid=True)
    plt.ylabel('Power (GW)')
    plt.title('Power data in week starting on day ' + str(startDay) + ' of month ' + str(start.month) + ' in ' + str(start.year) + ' for ' + str(i))
    plt.close

#Plotting 24 hours of one day
def plotDay(year,month,day,zone):
    i=str(zone)
    hours=power['Hour']
    zone_col=power[[i]]
    zone_col['Rolling Mean'] = zone_col[i].rolling(window=30).mean()
    #zone_col['Max'] = zone_col[i].rolling(window=3).max()
    start=datetime(int(year),int(month),int(day))
    end=datetime(int(year),int(month),int(day))
    plt.plot(hours[(start<=zone_col.index) & (zone_col.index<=end)],zone_col[(start<=zone_col.index) & (zone_col.index<=end)])
    plt.ylabel('Power (GW)')
    plt.xlabel('Hour')
    plt.title('24H Power Data on ' + str(start.year) + '/' + str(start.month) + '/' + str(start.day) + ' for ' + str(i))
    plt.close

    
        
#plotAllZonesAllMonths(2016)      
#plotAllMonths(2016,'Toronto')
#plotWeek(2016, 'Toronto', 5, 16)
plotDay(2016,5,16,'Northwest')
```
